## 基本使用

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait


browser = webdriver.Chrome()
try:
    browser.get('https://www.baidu.com')
    input = browser.find_element_by_id('kw') #找到id为kw的元素
    input.send_keys('Python')  # 模拟输入
    input.send_keys(Keys.ENTER)# 模拟回车
    wait = WebDriverWait(browser, 10) #等待10秒
    wait.until(EC.presence_of_element_located((By.ID, 'content_left')))#等待content_left元素加载完成
    
    print(browser.current_url)
    print(browser.get_cookies())
    print(browser.page_source)
finally:
    browser.close()
```

## 查找元素

### 单个元素

```python
from selenium import webdriver

browser = webdriver.Chrome()
browser.get('https://www.taobao.com')

input = browser.find_element_by_id('q')  #找元素
print(input)
browser.close()
```

- find_element_by_id
- find_element_by_name
- find_element_by_link_text
- find_element_by_partial_link_text
- find_element_by_tag_name
- find_element_by_class_name
- find_element_by_css_selector

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

browser = webdriver.Chrome()
browser.get('https://www.taobao.com')

input = browser.find_element(By.ID,'q')  #通用的找元素方式
print(input)
browser.close()
```

### 多个元素

```python
from selenium import webdriver
browser = webdriver.Chrome()
browser.get('https://www.taobao.com')
lis = browser.find_elements_by_css_selector('.service-bd li')
print(lis)
browser.close()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
browser = webdriver.Chrome()
browser.get('https://www.taobao.com')
lis = browser.find_elements(By.CSS_SELECTOR,'.service-bd li')  #通过CSS选择器将servicebid下的li元素都获取
print(lis)
browser.close()
```

### 元素交互操作

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
browser = webdriver.Chrome()
browser.get('https://www.taobao.com')
input = browser.find_element(By.ID,'q') #找到id为q的元素,通过查看html源码能发现q元素对应的时搜索框
input.send_keys('iphone')  # 模拟输入
time.sleep(1)
input.clear()
input.send_keys('ipad')
btn = browser.find_element(By.CLASS_NAME,'btn-search')
btn.click()
```
